# Building GPT with Andrej Karpathy

```python
url = 'https://raw.githubusercontent.com/karpathy/char-rnn/master/data/tinyshakespeare/input.txt'
```

```python
import urllib.request
urllib.request.urlretrieve(url, 'input.txt')
```

```python
# Read the input.txt file and store it in a variable called text; this is the dataset we will train on
with open('input.txt', 'r', encoding='utf-8') as f:
    text = f.read()
```

```python
print("length of dataset in characters: ", len(text))
```

```python
print(text[:1000])  # first 1000 characters
```

## Building the Vocabulary

We need to know what characters exist in our dataset before we can map them. Here we extract the unique characters from the dataset and sort them, giving us our vocabulary and its size.

```python
chars = sorted(list(set(text)))  # unique characters in the dataset
# set() gives unique characters,
# list() converts it to a list,
# sorted() sorts them in ascending order

vocab_size = len(chars) # the number of unique characters in the dataset
```

```python
print("all the unique characters:", ''.join(chars))
```

```python
print("vocab size:", vocab_size)
```
